### Species Area Relationship(SAR) 
We consider a territory of area $A$ divided into $k$ subregions of area $a_k$ such that $A = \sum a_k$ and we define the adimentional quantity $\alpha_k = \frac{a_k}{A}$.Typically the total area is equally divided among all the subregions, so $\alpha_k = \alpha$ .

If $P(n)$ is log-series distribution 
$$
P(n_i) = \frac{(1-\nu)^{n_i}}{n_i} \frac{1}{|log(\nu)|}

$$
we get SAR as the following:
$$

SAR(a) = S \bigg(1-(\frac{log[\nu(1- \alpha)+ \alpha]}{log(\nu)})\bigg)

$$
$S$ is a total species on the whole territory and $\nu$ is the growth value of species.\
For the power-low SAR equation we have:
$$
S(a) = k a^{z}
$$
### Endemic Area Relationship (EAR)
ÙŒWe can define the Endemic Area Relationship as the number of species whose individuals are completely contained inside a given area $a$. The EAR equation is:
$$
EAR(a) = S \sum^{\infty}_{n_i=1} \alpha^{n_i}P(n_i)
$$
Again we can put $P(n_i)$ in the equation 
$$EAR(a)= S \sum^{\infty}_{n_i} \alpha^{n_i} \frac{(1-\nu)^{n_i}}{n_i} \frac{1}{|log \nu|}$$
$$EAR(a)= S \sum^{\infty}_{n_i}\frac{(\alpha(1-\nu))^{n_i}}{n_i}\frac{1}{|log\nu|}$$ 
and use the polynomial series :
$$\sum^{\infty}_{n=1}\frac{z^n}{n} = -log(1-z) $$
Imagine that $z= \alpha(1-\nu)$ and $n=n_i$ so the final equation is become:

$$
EAR(a) = -S \frac{log[1-\alpha(1-\nu)]}{|log(\nu)|}
$$


```python
# S : total population
# alpha : dimentional quantity alpha_k = a_k/A
# nu : birth rate
# A : area that devided to sub region


def P(n, nu):
    return ((1-nu)**n)/n*abs(np.log(nu))

def SAR(S, nu, alpha):
    return S*(1-(np.log(nu*(1- alpha)+ alpha)/np.log(nu)))

def EAR(S , alpha, nu):
    return -S*(np.log(1-alpha*(1-nu))/abs(np.log(nu)))

def power_low(k , alpha , z):
    return k* (alpha**z)
```

```python
alpha = np.linspace(0.01, 1, 100) # a

sar=SAR(S=100, nu=0.2, alpha=alpha)
ear=EAR(S=100 , alpha=alpha, nu=0.2)
power= power_low(k=100 , alpha=alpha , z=0.25)
plt.figure(figsize=(15,10))
plt.plot(alpha , sar, label='SAR(a)')
plt.plot(alpha, ear, label='EAR(a)')
plt.plot(alpha, power, label='Power Law SAR (z=0.25)')
plt.title('growth rate = 0.2', fontsize = 20)
plt.xlabel('')
plt.ylabel("population")
plt.legend(loc='best')
plt.grid()
```
